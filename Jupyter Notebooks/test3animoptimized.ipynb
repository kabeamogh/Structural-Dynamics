```python
from ipycanvas import Canvas
from math import pi
import numpy as np
from typing import Tuple
from ipycanvas import hold_canvas
import asyncio


canvas = Canvas(width=600, height=600, sync_image_data=True)
display(canvas)

t = np.linspace(0, 2, 200)

async def animate_both():
    for j in range(len(t)):
        with hold_canvas():
            canvas.clear()
    
            def draw_oval(canvas, center: Tuple[float, float], a: float, b: float, fill_style: str = "blue", stroke_style: str = "black", line_width: float = 1.0):
                x, y = center
                with hold_canvas():
                    canvas.save()
                    canvas.translate(x, y)
                    canvas.scale(a / b, 1)  
                    canvas.begin_path()
                    canvas.arc(0, 0, b, 0, 2 * np.pi)
                    canvas.fill_style = fill_style
                    canvas.fill()
                    canvas.stroke_style = stroke_style
                    canvas.stroke()
                    canvas.line_width = line_width
                    canvas.restore()
            
            #Center Line
            canvas.set_line_dash([5, 5]) 
            canvas.stroke_line(150, 350, 480, 350) 
            canvas.set_line_dash([]) 
            canvas.stroke_line
            
            #Horizontal lines outside rectangle
            canvas.line_width = 3
            canvas.stroke_line(170, 350, 120, 350)
            canvas.stroke_line(460, 350, 510, 350)
            
            #The barbs above and below the horizontal lines
            canvas.line_width = 2
            canvas.stroke_line(160, 340, 125, 340)
            canvas.stroke_line(135, 335, 130, 340)
            canvas.stroke_line(145, 335, 140, 340)
            canvas.stroke_line(155, 335, 150, 340)
            
            canvas.stroke_line(470, 340, 505, 340)
            canvas.stroke_line(475, 340, 480, 335)
            canvas.stroke_line(485, 340, 490, 335)
            canvas.stroke_line(495, 340, 500, 335)
            
            canvas.stroke_line(160, 360, 125, 360)
            canvas.stroke_line(130, 365, 135, 360)
            canvas.stroke_line(140, 365, 145, 360)
            canvas.stroke_line(150, 365, 155, 360)
            
            canvas.stroke_line(470, 360, 505, 360)
            canvas.stroke_line(480, 360, 475, 365)
            canvas.stroke_line(490, 360, 485, 365)
            canvas.stroke_line(500, 360, 495, 365)
            
            #Double arrow on the center line
            canvas.line_width = 3
            canvas.stroke_line(370, 350, 420, 350)
            canvas.stroke_line(415, 345, 420, 350)
            canvas.stroke_line(415, 355, 420, 350)
            canvas.stroke_line(405, 345, 410, 350)
            canvas.stroke_line(405, 355, 410, 350)
            
            
            #Lines above the rectangle
            canvas.stroke_style = 'black'
            canvas.line_width = 2
            canvas.stroke_line(200, 270, 200, 295)  
            canvas.stroke_line(195, 280, 200, 275)
            canvas.stroke_line(195, 285, 200, 280)
            canvas.stroke_line(195, 290, 200, 285)
            
            canvas.font = "32px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("Φ", 210, 275)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("1", 232, 275)
            
            canvas.stroke_line(200, 285, 255, 285) 
            canvas.stroke_line(250, 280, 255, 285)
            canvas.stroke_line(250, 290, 255, 285)  
            
            canvas.stroke_line(245, 280, 250, 285)
            canvas.stroke_line(245, 290, 250, 285)  
            
            canvas.stroke_style = 'black'
            canvas.line_width = 2
            canvas.stroke_line(350, 270, 350, 295)  
            canvas.stroke_line(345, 280, 350, 275) 
            canvas.stroke_line(345, 285, 350, 280) 
            canvas.stroke_line(345, 290, 350, 285) 
            
            canvas.font = "32px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("Φ", 360, 275)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("1", 382, 275)
            
            canvas.stroke_line(350, 285, 400, 285) 
            canvas.stroke_line(395, 280, 400, 285)
            canvas.stroke_line(395, 290, 400, 285)  
            
            canvas.stroke_line(390, 280, 395, 285)
            canvas.stroke_line(390, 290, 395, 285)  
            
            #Notations around the rectangle
            canvas.font = "32px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("J", 370, 430)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("S2", 382, 440)
            
            canvas.font = "32px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("J", 220, 430)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("S1", 232, 440)
            
            canvas.font = "32px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("c", 300, 320)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("Φ", 312, 330)
            
            canvas.line_width = 1
            
            
            canvas.fill_style = 'black'
            canvas.fill_rect(220, 300, 30, 20)
            canvas.fill_style = 'black'
            canvas.fill_rect(220, 380, 30, 20)
            mass_center = [220.0, 350.0]
            draw_oval(canvas, mass_center, a=20.0, b=50.0, fill_style="black", stroke_style="black", line_width=1.0)
            mass_center = [250.0, 350.0]
            draw_oval(canvas, mass_center, a=20.0, b=50.0, fill_style="lightgrey", stroke_style="black", line_width=1.0)
            
            
            big_circle_r1 = 13
            big_circle_r2 = 28
            angle = 7 * t[j]
            dot_x = 250 + big_circle_r1 * np.cos(angle)
            dot_y = 350 + big_circle_r2 * np.sin(angle)
            canvas.fill_circle(dot_x, dot_y, 5)
                     
            #Main Blocks
            canvas.fill_style = 'lightgrey'
            canvas.fill_rect(255, 340, 95, 20)
            canvas.stroke_style = 'lightgrey'
            canvas.fill_style = 'black'
            canvas.stroke_rect(255, 340, 95, 20)
            canvas.stroke_style = 'black'
            canvas.stroke_line(260, 340, 350, 340)
            canvas.stroke_line(260, 360, 350, 360)
            canvas.stroke_line(350, 340, 350, 360)
            
            
            canvas.begin_path()
            canvas.move_to(260,340)
            canvas.stroke_style = 'black'
            canvas.arc(260, 350, 10, 3* pi/2, pi/2, True)
            canvas.stroke()

            canvas.fill_style = 'black'
            canvas.fill_rect(360, 300, 30, 20)
            canvas.fill_style = 'black'
            canvas.fill_rect(360, 380, 30, 20)
            mass_center = [360.0, 350.0]
            draw_oval(canvas, mass_center, a=20.0, b=50.0, fill_style="black", stroke_style="black", line_width=1.0)
            mass_center = [390.0, 350.0]
            draw_oval(canvas, mass_center, a=20.0, b=50.0, fill_style="lightgrey", stroke_style="black", line_width=1.0)
            
            big_circle_rr1 = 13
            big_circle_rr2 = 28
            angle = 2 * np.pi * t[j]
            dot_x = 390 + big_circle_rr1 * np.sin(angle)
            dot_y = 350 + big_circle_rr2 * np.cos(angle)
            canvas.fill_circle(dot_x, dot_y, 5)

            #Double arrow on the center line
            canvas.line_width = 3
            canvas.stroke_line(390, 350, 420, 350)
            canvas.stroke_line(415, 345, 420, 350)
            canvas.stroke_line(415, 355, 420, 350)
            canvas.stroke_line(405, 345, 410, 350)
            canvas.stroke_line(405, 355, 410, 350)
            
            canvas.font = "17px serif"
            canvas.fill_style = 'black'
            canvas.stroke_style = 'black'
            canvas.fill_text("M(t)", 415, 335)

        await asyncio.sleep(0.02)
    
await animate_both()    
        
```
