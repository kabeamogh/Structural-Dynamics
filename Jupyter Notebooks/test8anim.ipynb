```python
from ipycanvas import Canvas, hold_canvas, MultiCanvas
from math import pi
import time
import numpy as np
from src.utils.helper import ghetto_feder_daempfer_element_top
from src.utils.helper import ghetto_feder_daempfer_element_bottom
from src.utils.ext_utils import spring_module
from IPython.display import HTML, display

# Remove Jupyter padding
HTML('<style>.canvas-container { margin: 0; padding: 0; }</style>')

# Create canvas
canvas = Canvas(width=800, height=800, sync_image_data=True)
display(canvas)

# Animation parameters
dt = 0.01  # Time step (s)
t_max = 2.0  # Total time (s)
t = np.arange(0, t_max, dt)

# Lines for rectangles
x_lines_8_1 = np.linspace(250 + 130/21, 250 + 20*130/21, 20)  # Aufgabe 8.1
x1_lines_8_1 = np.linspace(410 + 130/21, 410 + 20*130/21, 20)


# Animation loop
for i in range(len(t)):
    with hold_canvas():
        canvas.clear()
        canvas.fill_style = 'white'
        canvas.fill_rect(0, 0, 800, 800)

        # Motion offsets
        mx1 = 380 + 30 * np.sin(2 * np.pi * t[i])
        mx2 = 540 + 30 * 2 * np.sin(2 * np.pi * t[i])
        

        # Aufgabe 8.1: Elongate and deelongate rectangle with particle compression
        width_8_1 = float(130 + 30 * np.sin(2 * np.pi * t[i]))  # 110 to 150 pixels
        x_lines_new_8_1 = 250 + (x_lines_8_1 - 250) * (width_8_1 / 130)

        
        canvas.fill_style = 'white'
        canvas.fill_rect(250, 150, width_8_1, 20)
        canvas.stroke_style = 'black'
        canvas.stroke_rect(250, 150, width_8_1, 20)
        for x in x_lines_new_8_1:
            canvas.stroke_line(float(x), 150, float(x), 170)

        
        x1_lines_new_8_1 = (mx1 + 30) + (x1_lines_8_1 - 410) * (width_8_1 / 130)

        # Static right rectangle and vertical blocks
        canvas.fill_style = 'white'
        canvas.fill_rect(mx1 + 30, 150, width_8_1, 20)
        canvas.stroke_style = 'black'
        canvas.stroke_rect(mx1 + 30, 150, width_8_1, 20)
        for x1 in x1_lines_new_8_1:
            canvas.stroke_line(float(x1), 150, float(x1), 170)



        canvas.fill_style = 'lightgrey'
        canvas.fill_rect(mx1, 100, 30, 120)
        canvas.stroke_rect(mx1, 100, 30, 120)
        canvas.fill_style = 'lightgrey'
        canvas.fill_rect(mx2, 100, 30, 120)
        canvas.stroke_rect(mx2, 100, 30, 120)

        canvas.line_width = 2
        canvas.stroke_line(250, 135, 250, 185)
        canvas.stroke_line(250, 139, 240, 134)
        canvas.stroke_line(250, 145, 240, 140)
        canvas.stroke_line(250, 151, 240, 146)
        canvas.stroke_line(250, 157, 240, 152)
        canvas.stroke_line(250, 163, 240, 158)
        canvas.stroke_line(250, 169, 240, 164)
        canvas.stroke_line(250, 175, 240, 170)
        canvas.stroke_line(250, 181, 240, 176)

        canvas.font = "22px serif"
        canvas.fill_style = 'black'
        canvas.fill_text("ﾏ！  ,GI  ,l", 280, 130)
        canvas.fill_text("ﾏ！  ,GI  ,l", 440, 130)
        canvas.fill_text("J", 390, 80)
        canvas.fill_text("J", 550, 80)

        canvas.font = "17px serif"
        canvas.fill_text("p", 299, 130)
        canvas.fill_text("p", 459, 130)
        canvas.fill_text("T", 339, 134)
        canvas.fill_text("T", 499, 134)
        canvas.fill_text("1", 400, 90)
        canvas.fill_text("2", 560, 90)


#################################################################################

        # Lines for rectangles
        x2_lines_8_2 = np.linspace(160 + 150/21, 160 + 22*150/22, 21)  # Aufgabe 8.2
        x3_lines_8_2 = np.linspace(570 - 150/21, 570 - 22*150/22, 21)

        # Aufgabe 8.2: Spring compression with rectangle elongation and opposing particle movement
        motionright = 20 * np.sin(2 * np.pi * t[i])
        motionleft = - 20 * np.sin(2 * np.pi * t[i])
        width_8_2_1 = 310 + 20 * np.sin(2 * np.pi * t[i]) # 130 to 170 pixels
        width_8_2_11 = -(150 - 20 * 2 * np.sin(2 * np.pi * t[i]))
        width_8_2_2 = 420 - 20 * np.sin(2 * np.pi * t[i]) 
        width_8_2_22 = 150 - 20 * 2 * np.sin(2 * np.pi * t[i])
        
        width_8_2 = float(150 + 30 * np.sin(2 * np.pi * t[i]))
        x2_lines_new = np.copy(x2_lines_8_2) + motionleft
        x2_lines_new[20] = width_8_2_1  # First particle (rightmost) moves with the right edge
        for j in range(19, -1, -1):  # Adjust other particles from right to left
            x2_lines_new[j] = x2_lines_new[j + 1] - ((x2_lines_8_2[j + 1] - x2_lines_8_2[j]) * (abs(width_8_2_11) / 150))

        x3_lines_new = np.copy(x3_lines_8_2) + motionright
        x3_lines_new[20] = width_8_2_2
        for j in range(19, -1, -1):  # Right rectangle movement
            x3_lines_new[j] = x3_lines_new[j + 1] - ((x3_lines_8_2[j + 1] - x3_lines_8_2[j]) * (abs(width_8_2_11) / 150))


        canvas.fill_style = 'lightgrey'
        canvas.fill_rect(width_8_2_1, 485, width_8_2_11, 15)
        canvas.stroke_rect(width_8_2_1, 485, width_8_2_11, 15)
        for x in x2_lines_new:
            canvas.stroke_line(float(x), 485, float(x), 500)

        canvas.fill_style = 'lightgrey'
        canvas.fill_rect(width_8_2_2, 485, width_8_2_22, 15)
        canvas.stroke_rect(width_8_2_2, 485, width_8_2_22, 15)
        for x1 in x3_lines_new:
            canvas.stroke_line(float(x1), 485, float(x1), 500)

        # Draw barbed lines and triangles

        # 1st Triangle
        canvas.stroke_polygon([(163 + (3 * motionright), 500), (153 + (3 * motionright), 515), (173 + (3 * motionright), 515)])
        canvas.stroke_line(150, 520, 170, 520)
        canvas.stroke_line(152, 520, 158, 525)
        canvas.stroke_line(158, 520, 164, 525)
        canvas.stroke_line(164, 520, 170, 525)

        # 4th Triangle
        canvas.stroke_polygon([(567 + (3 * motionleft), 500), (557 + (3 * motionleft), 515), (577 + (3 * motionleft), 515)])
        canvas.stroke_line(560, 520, 580, 520)
        canvas.stroke_line(562, 520, 568, 525)
        canvas.stroke_line(568, 520, 574, 525)
        canvas.stroke_line(574, 520, 580, 525)

        # 2nd Triangle
        canvas.stroke_polygon([(310 + motionright, 500), (300 + motionright, 515), (320 + motionright, 515)])
        canvas.stroke_line(300, 520, 320, 520)
        canvas.stroke_line(302, 520, 308, 525)
        canvas.stroke_line(308, 520, 314, 525)
        canvas.stroke_line(314, 520, 320, 525)
        
        # 3rd Triangle
        canvas.stroke_polygon([(420 + motionleft, 500), (410 + motionleft, 515), (430 + motionleft, 515)])
        canvas.stroke_line(410, 520, 430, 520)
        canvas.stroke_line(412, 520, 418, 525)
        canvas.stroke_line(418, 520, 424, 525)
        canvas.stroke_line(424, 520, 430, 525)

        # Slant lines
        canvas.stroke_line(200, 460, 220, 475)
        canvas.stroke_line(500, 460, 480, 475)

        # Spring compression from both sides
        left_spring_end = 310 + 20 * np.sin(2 * np.pi * t[i])
        right_spring_end = 420 - 20 * np.sin(2 * np.pi * t[i])
        x_coords, y_coords = spring_module.spring(start=(left_spring_end, 492), end=(right_spring_end, 492), nodes=7, width=25)
        x_y_pairs = list(zip(x_coords, y_coords))
        canvas.stroke_lines(x_y_pairs)

        canvas.fill_style = 'black'
        canvas.font = "22px serif"
        canvas.fill_text("l, EA, ﾏ、", 170, 450)
        canvas.fill_text("l, EA, ﾏ、", 470, 450)
        canvas.fill_text("k", 360, 470)

    time.sleep(0.02)  # ~50 FPS
```
